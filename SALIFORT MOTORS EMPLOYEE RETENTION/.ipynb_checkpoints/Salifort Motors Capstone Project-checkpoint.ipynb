#### Modeling (Logistic Regression)

**Logistic Binomial Regression Modeling**\
Data Cleaning and Preparation

```python
## Encode categorical values to category datatype
### Create new DataFrame for encoding
df_enc = df1.copy()
### Convert columns to Categorical
df_enc['department'] = df_enc['department'].astype('category')
#df_enc['salary'] = pd.Categorical(df_enc['salary'],categories=['low', 'medium', 'high'], ordered = True)
df_enc['salary'] = (df_enc['salary'].astype('category').cat.set_categories(['low', 'medium', 'high']).cat.codes)
### Encode
df_enc = pd.get_dummies(df_enc, drop_first = False)
```

```python
## Check linear correlation using a heatmap
plt.figure(figsize = (8,4))
sns.heatmap(data = df_enc[['satisfaction_level','last_evaluation','number_project','average_monthly_hours','tenure']].corr(), 
            vmin = -1, vmax = 1, annot = True, cmap=sns.color_palette("vlag", as_cmap=True))
plt.show()
```

```python
#Remove previously determined outliers due to sensitivity of the model.
percentile25 = df_enc['tenure'].quantile(0.25)
percentile75 = df_enc['tenure'].quantile(0.75)

iqr = percentile75 - percentile25
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr

df_enc_logreg = df_enc[(df_enc['tenure'] <= upper_limit) & (df_enc['tenure'] >= lower_limit)]
```

```python
#Isolation of dependent variable
y = df_enc_logreg['left']
```

```python
#Isolation of independent variable
X = df_enc_logreg.drop('left', axis = 1)
```

```python
#Check Balance of Dependent Variable
df_enc_logreg['left'].value_counts()
df_enc_logreg['left'].value_counts(normalize=True)
```

**Logistic Binomial Regression Modeling**\
Modeling Fitting

```python
#Creation of Train and Test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42, stratify=y)
```

```python
#Fitting into Binary Logistic Regression
log_clf = LogisticRegression(random_state = 42, max_iter=500).fit(X_train, y_train)
```

```python
#Present Coefficient of the Model
col_names = X_train.columns
coef_array = []
for i in range(len(X_train.columns)):
    x = log_clf.coef_[0][i]
    coef_array.append(x)
coef = pd.DataFrame(data={'Coefficient':coef_array}, index = X_train.columns)
coef['Coefficient'] = coef['Coefficient'].apply(lambda x: '{:,.5f}'.format(x))
coef.sort_values('Coefficient', ascending = False)
```

```python
#Identify the intercept
log_clf.intercept_
```

**Logistic Binomial Regression Modeling**\
Model Evaluation

```python
#Get predicted dependent Variables
y_pred = log_clf.predict(X_test)
```

```python
#Create Confusion Matrix
cm = confusion_matrix(y_test, y_pred, labels = log_clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels= log_clf.classes_)
disp.plot(values_format='', cmap = 'Blues')
plt.show()
```

```python
#Calculate Evaluation Matrics
target_names = ['Predicted would not leave', 'Predicted would leave']
print(classification_report(y_test, y_pred, target_names=target_names))

model_name = 'Logistic Regression'
accuracy = metrics.accuracy_score(y_test, y_pred)
precision = metrics.precision_score(y_test, y_pred, average='weighted')
recall = metrics.recall_score(y_test, y_pred, average='weighted')
f1 = metrics.f1_score(y_test, y_pred, average='weighted')
auc = metrics.roc_auc_score(y_test, y_pred, average='weighted')

Log_reg = pd.DataFrame()
Log_reg = pd.DataFrame({'model': [model_name],
                      'precision': [precision],
                      'recall': [recall],
                      'f1': [f1],
                      'accuracy': [accuracy],
                      'AUC': [auc]
                     })
Log_reg
```

**Helper Function**: Results Summary

```python
def make_results(model_name:str, model_object, metric:str):
    '''
    Arguments:
        model_name (string): what you want the model to be called in the output table
        model_object: a fit GridSearchCV object
        metric (string): precision, recall, f1, accuracy, or auc
  
    Returns a pandas df with the F1, recall, precision, accuracy, and auc scores
    for the model with the best mean 'metric' score across all validation folds.  
    '''

    # Create dictionary that maps input metric to actual metric name in GridSearchCV
    metric_dict = {'auc': 'mean_test_roc_auc',
                   'precision': 'mean_test_precision',
                   'recall': 'mean_test_recall',
                   'f1': 'mean_test_f1',
                   'accuracy': 'mean_test_accuracy'
                  }

    # Get all the results from the CV and put them in a df
    cv_results = pd.DataFrame(model_object.cv_results_)

    # Isolate the row of the df with the max(metric) score
    best_estimator_results = cv_results.iloc[cv_results[metric_dict[metric]].idxmax(), :]

    # Extract Accuracy, precision, recall, and f1 score from that row
    auc = best_estimator_results.mean_test_roc_auc
    f1 = best_estimator_results.mean_test_f1
    recall = best_estimator_results.mean_test_recall
    precision = best_estimator_results.mean_test_precision
    accuracy = best_estimator_results.mean_test_accuracy
  
    # Create table of results
    table = pd.DataFrame()
    table = pd.DataFrame({'model': [model_name],
                          'precision': [precision],
                          'recall': [recall],
                          'F1': [f1],
                          'accuracy': [accuracy],
                          'auc': [auc]
                        })
  
    return table
```

**Helper Function**: Pickle Model

```python
# Define a path to the folder where you want to save the model
path = ''
```

```python
def write_pickle(path, model_object, save_as:str):
    '''
    In: 
        path:         path of folder where you want to save the pickle
        model_object: a model you want to pickle
        save_as:      filename for how you want to save the model

    Out: A call to pickle the model in the folder indicated
    '''    

    with open(path + save_as + '.pickle', 'wb') as to_write:
        pickle.dump(model_object, to_write)
```

```python
def read_pickle(path, saved_model_name:str):
    '''
    In: 
        path:             path to folder where you want to read from
        saved_model_name: filename of pickled model you want to read in

    Out: 
        model: the pickled model 
    '''
    with open(path + saved_model_name + '.pickle', 'rb') as to_read:
        model = pickle.load(to_read)

    return model
```

#### Model Evaluation with Test Subset

```python
#Helper Function to get Results
def get_scores(model_name:str, model, X_test_data, y_test_data):
    '''
    Generate a table of test scores.

    In: 
        model_name (string):  How you want your model to be named in the output table
        model:                A fit GridSearchCV object
        X_test_data:          numpy array of X_test data
        y_test_data:          numpy array of y_test data

    Out: pandas df of precision, recall, f1, accuracy, and AUC scores for your model
    '''

    preds = model.best_estimator_.predict(X_test_data)

    auc = metrics.roc_auc_score(y_test_data, preds)
    accuracy = metrics.accuracy_score(y_test_data, preds)
    precision = metrics.precision_score(y_test_data, preds)
    recall = metrics.recall_score(y_test_data, preds)
    f1 = metrics.f1_score(y_test_data, preds)

    table = pd.DataFrame({'model': [model_name],
                          'precision': [precision], 
                          'recall': [recall],
                          'f1': [f1],
                          'accuracy': [accuracy],
                          'AUC': [auc]
                         })
  
    return table
```
